```python
# combine videos of predicted probability fields for each well

import os
import re
import subprocess

def extract_start_number(filename):
    match = re.search(r'F(\d+)_', filename)
    return int(match.group(1)) if match else 0

def combine_avi_videos(video_folder, output_filename):
    
    avi_files = [file for file in os.listdir(video_folder) if file.endswith('.avi') and 'file' in file]
    sorted_avi_files = sorted(avi_files, key=extract_start_number)
    print(sorted_avi_files)
    
    with open('file_list.txt', 'w') as f:
        for file in sorted_avi_files:
            f.write(f"file '{os.path.join(video_folder, file)}'\n")

    subprocess.run([
        'ffmpeg', '-f', 'concat', '-safe', '0',
        '-i', 'file_list.txt', '-c', 'copy', output_filename
    ], check=True)
    
    os.remove('file_list.txt')



for row in "ABCDEFGH":
    for col in range(1, 13):
        well = f"{row}{col}"
        
        well_path = f"/path/to/folders/{well}"   # path to the folders containing the videos of each well

        if os.path.isdir(well_path) and os.listdir(well_path):  # Check if folder exists and is not empty
            combine_avi_videos(well_path, f"{well_path}/combined_video_name_{well}.avi")  # change file name
            print(f"{well} is done")

```
